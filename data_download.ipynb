```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Function to display sample images
def display_sample_images(image_dir, num_samples=3):
    image_files = []
    for root, _, files in os.walk(image_dir):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                image_files.append(os.path.join(root, file))

    if not image_files:
        print(f"No image files found in {image_dir}")
        return

    # Randomly select images
    selected_images = pd.Series(image_files).sample(min(num_samples, len(image_files))).tolist()

    plt.figure(figsize=(15, 5))
    for i, img_path in enumerate(selected_images):
        ax = plt.subplot(1, num_samples, i + 1)
        img = mpimg.imread(img_path)
        plt.imshow(img, cmap='gray')
        plt.title(os.path.basename(os.path.dirname(img_path)))
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Display sample images from the training set's 'NORMAL' and 'PNEUMONIA' subdirectories
# Corrected paths to include the 'chest_xray' subdirectory
train_normal_dir = os.path.join(path, 'chest_xray', 'train', 'NORMAL')
train_pneumonia_dir = os.path.join(path, 'chest_xray', 'train', 'PNEUMONIA')

print("\nSample NORMAL images from training set:")
display_sample_images(train_normal_dir)

print("\nSample PNEUMONIA images from training set:")
display_sample_images(train_pneumonia_dir)
```
